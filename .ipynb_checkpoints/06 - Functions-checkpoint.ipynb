- A function is a set of instructions that can be "called"/"invoked" whenever needed.
- Once you've defined a function, you can use it multiple times
- This reduces the effort of rewriting the same logic again and again

````python
def make_tea():
    print("Boil some water")
    print("Put ginger, sugar, tea leaves in it")
    print("Put some milk in it")
````

````python
# def = define
````

````python
make_tea() # calling/invoking the function
````

````python
make_tea()
make_tea()
make_tea()
````

````python
for i in range(5):
    make_tea()
````

### Positional vs Keyword Arguments

````python
def random(a, b, c, d):
    print(a, b, c, d)
````

````python
random(5, 2, 1, 9)
# positional arguments
````

````python
random(b=2, d=1, a=5, c=4)
# keyword arguments
````

````python
random(7, 8, d=0, c=3)
# positional => Keyword
````

````python
random(5, 7, b=1, d=9)
# a = 5
# b = 7
# again b = 1 (error)

# positional => keyword
````

````python
random(a=1, b=5, 10, 4)
# keyword => positional
````

````python
# Rule: We cannot have positional arguments after keyword arguments
````

### Scope of variables

````python
a = 10 # Global variable

def random():
    a = 50 # Local variable
    print("Inside the function", a)
    
random()
print("Outside the function", a)
````

````python
a = 10 # Global variable

def praneeth():
    global a
    # When you see 'a' in the future statements,
    # don't declare a new local variable
    # Simply use the global 'a' variable only
    
    a = 50 # I want to update the global variable only
    print("Inside the function", a)
    
praneeth()
print("Outside the function", a)
````

````python
a = 10

def random():
    print("Inside the function", a)
    
random()
print("Outside the function", a)
````

````python
def main():
    tc = int(input())
    while tc > 0:
        tc -= 1
        num = int(input())
        rev_num = 0
        while num > 0:
            rev_num = (rev_num * 10) + (num % 10)
            num = num // 10
        print(rev_num)
    
    return 0

main()
````
